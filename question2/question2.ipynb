### General Setup:

```python
#Importing the required modules
import math as mt
import numpy as np
import matplotlib.pyplot as plt

#Defining a function to compute factorials (defined for non-negative x)
def factorial(x):
    if x == 0:
        return 1
    else:
        return mt.factorial(x)
```

### Section a)

```python
#Define the binomial coefficient (defined for x,y >= 0 and x >= y)
def binomialcoefficient(x,y):
        return int(factorial(x)/(factorial(y)*factorial(x-y)))
    
#Checking if the result for y=0 is 1 and if the output is an interger
print('Result for arbitrary x and y=0: ', binomialcoefficient(5, 0))
print('Checking if the result in an integer: ', type(binomialcoefficient(5, 0)))
```

### Section b)

```python
#Use the binomial cofficient function in a loop to print pascal's triangle
for i in range(20):
    values = []
    for j in range(i+1):
        values.append(binomialcoefficient(i,j))
    print("Line " + str(i+1) + ":", values)
```

### Section c)

```python
#Choosing some arbitrary parameters for p, n and k
p = 0.39
n = 7
k = 3

#Defining the probability function for the coin toss
def coinprobability(p, n, k):
        return (binomialcoefficient(n,k)*(p**k)*((1-p))**(n-k))

#Printing the probability as a percentage for chosen p,n and k
percentchance = 100*coinprobability(p, n, k)
print('Probability of obtaining heads ' + str(k) + ' times for ' + str(n) + ' flips: ' + str(percentchance) + "%")
```

### Section d)

```python
#Define a function for that gives the number of successful trails over N trials
def Ntrialexperiment(N):
    trialsuccess = 0
    for i in range(int(N)):
        heads = 0
        tails = 0
        for j in range(n):
            coinflip = (np.random.random_sample())
            if coinflip <= p:
                heads = heads + 1
            else:
                tails = tails + 1
        if heads >= k:
            trialsuccess = trialsuccess + 1
    return 100*(trialsuccess/N)

#Take N in [1, 500] 
x = np.arange(1, 500)
vectorized = np.vectorize(Ntrialexperiment)
y = vectorized(x)

#Setting up the plot for N in [1, 500] 
plt.figure(figsize=(10,8))
plt.plot(x, y, '-')
plt.xlabel('Number of Trials', fontsize=10)
plt.ylabel('Percent Trial Success', fontsize=10)
plt.title('Simulated Experiment Results for N Trials', fontsize=15)
plt.show()
```
